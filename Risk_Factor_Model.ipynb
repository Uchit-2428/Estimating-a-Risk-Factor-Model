Importing necessary libraries

Loading the famafrench dataset

F-F_Research_Data_Factors is the famafrench 3 factor model dataset

F-F_Momentum_Factor contains the momentum factor data which is added to the 3 factor model to make the famafrench 4 factor model

```python
start = '1926-01-01'
```

```python
ff = pandas_datareader.famafrench.FamaFrenchReader('F-F_Research_Data_Factors', freq='M', start=start).read()
```

```python
ff
```

```python
ff_df = ff[0]
```

Mkt-RF => Market Risk Premium
It’s the excess return of the market portfolio over the risk-free rate.

SMB => Small Minus Big
Measures the size premium — the historical tendency for small-cap stocks to outperform large-cap stocks.

HML => High Minus Low
Measures the value premium — the tendency for value stocks (high book-to-market ratios) to outperform growth stocks (low book-to-market ratios).

RF => Risk-Free Rate
It’s the monthly return of a risk-free asset.

Merging both the dataframes to create famafench 4 factor model dataframe


```python
import pandas as pd
ffac_merged_df = pd.merge(ff_df, ff_mom_df, on='Date', how='inner', sort=True, copy=True, indicator=False, validate='one_to_one')
```

Adding new column to both the dataframes to be used as a common identifier for merging the datasets


```python
AAPL_monthly_returns['str_date'] = AAPL_monthly_returns.index.astype(str)
AAPL_monthly_returns['dt_date'] = pd.to_datetime(AAPL_monthly_returns['str_date']).dt.strftime('%Y-%m')
```

```python
ffac_merged_df['str_date'] = ffac_merged_df.index.astype(str)
ffac_merged_df['dt_date'] = pd.to_datetime(ffac_merged_df['str_date']).dt.strftime('%Y-%m')
```

```python
AAPL_monthly_returns
```

```python
ffac_merged_df
```

Merging the APPL stock dataframe and the framafrench 4 factor model dataframe

```python
AAPL_ffac_merged_df = pd.merge(AAPL_monthly_returns, ffac_merged_df, on='dt_date', how='inner', sort=True, copy=True, indicator=False, validate='one_to_one')
```

```python
AAPL_ffac_merged_df
```

```python
AAPL_ffac_merged_df.drop(columns=['str_date_x', 'str_date_y'], inplace=True)
```

Adding a column which is the AAPL risk free rate which calculates the AAPL monthly excess returns


```python
AAPL_ffac_merged_df['AAPL_RF'] = AAPL_ffac_merged_df['AAPL']*100 - AAPL_ffac_merged_df['RF']
```

```python
AAPL_ffac_merged_df
```

```python
AAPL_ffac_merged_df.dropna(axis=0, inplace=True)
```

Using the OLS method of statsmodels package

OLS (Ordinary Least Squares) is a method to estimate the coefficients of a linear regression model by minimizing the sum of squared differences between observed and predicted values.


```python
from statsmodels.api import OLS
```

```python
list(AAPL_ffac_merged_df)
```

```python
AAPL_ffac_merged_df.rename(columns={'Mom   ': 'MOM'}, inplace=True)
```

```python
results = OLS(AAPL_ffac_merged_df['AAPL_RF'], AAPL_ffac_merged_df[['Mkt-RF', 'SMB', 'HML', 'MOM']], missing='drop').fit()
```

```python
results.summary()
```

Adding a column of constant (ones) to get the intercept of the Regression


```python
import statsmodels.tools
AAPL_ffac_merged_df_c = statsmodels.tools.add_constant(AAPL_ffac_merged_df, prepend=True)
```

```python
AAPL_ffac_merged_df_c
```

```python
results = OLS(AAPL_ffac_merged_df_c['AAPL_RF'], AAPL_ffac_merged_df_c[['const','Mkt-RF', 'SMB', 'HML', 'MOM']], missing='drop').fit()
```

```python
results.summary()
```

Final Inference:


*   The Fama-French 4-Factor model explains approximately 28.8% of Apple's excess returns, indicating a moderate fit.

*  The model is statistically significant overall (p < 0.001), with strong positive exposure to market risk and strong negative exposure to the value factor.

*   SMB (size) and MOM (momentum) factors are not statistically significant for Apple.

*   The results confirm that Apple behaves like a high-beta growth stock with considerable Stock-specific risk remaining unexplained by common factors.
